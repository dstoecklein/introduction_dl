# Multilayer Perceptron
- Simplest Deep neural network
- Multiple layers each fully connected (Every input influences every neuron)
- Overcome linearity by introducing one or more hidden layers
- Single hidden Layer: Number of Layers are 2 since we need to calculate hidden layer + output layer 

## Activation functions
- decider whether a neuron should fire

```python
import torch
from matplotlib import pyplot as plt
%matplotlib inline
```

### ReLu (Rectified linear unit)
- Given an element x, it outputs the maximum of that element starting at 0
- ignores negative

```python
x = torch.arange(-8.0, 8.0, 0.1, requires_grad=True)
y = torch.relu(x)
plt.plot(x.detach(), y.detach(), 'x', 'relu(x)')
```

### Sigmoid
- transforms input to output in interval (0,1) (squashing)

```python
y = torch.sigmoid(x)
plt.plot(x.detach(), y.detach(), 'x', 'sigmoid(x)')
```

### Tanh
- squashing input to interval (-1,1)

```python
y = torch.tanh(x)
plt.plot(x.detach(), y.detach(), 'x', 'sigmoid(x)')
```
