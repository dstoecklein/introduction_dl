```python
trainset = tf.keras.datasets.cifar10
```

```python
(x_train, y_train), (x_test, y_test) = trainset.load_data()
```

```python
y_train = y_train.flatten() 
y_test = y_test.flatten()
```

```python
x_train = x_train / 255.0 
x_test = x_test / 255.0
```

# 2. Model building

```python
K = 10 # number of classes
K
```

```python
model = tf.keras.models.Sequential()
model.add(layers.Conv2D(filters=6, kernel_size=5, activation='relu', input_shape=x_train[0].shape))
model.add(layers.MaxPooling2D(pool_size=2, strides=2))
model.add(layers.Conv2D(filters=16, kernel_size=5, activation='relu'))
model.add(layers.MaxPooling2D(pool_size=2, strides=2))
model.add(layers.Flatten())
model.add(layers.Dense(units=120, activation='relu'))
model.add(layers.Dense(units=84, activation='relu'))
model.add(layers.Dense(units=K))
```

```python
tf.keras.utils.plot_model(model, 'dnn_model.png', show_shapes=True, rankdir='LR')
```

```python
model.summary(
    line_length=None,
    positions=None,
    print_fn=None,
    expand_nested=False,
    show_trainable=False,
)
```

- `CategoricalCrossentropy`: Use this crossentropy loss function when there are two or more label classes. We expect labels to be provided in a one_hot representation. 
- `SparseCategoricalCrossentropy`: Use this crossentropy loss function when there are two or more label classes. We expect labels to be provided as integers.

```python
model.compile(
    optimizer=tf.keras.optimizers.SGD(learning_rate=0.001, momentum=0.9),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['acc']
)
```

# 3. Model training

```python
history = model.fit(x_train, y_train, batch_size=4, epochs=2)
```

# 4. Model evaluation

```python
test_loss, test_acc = model.evaluate(x_test, y_test) # test metrics
```

```python
plt.plot(history.history['acc'], label='accuracy')
#plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.ylim([0.5, 1])
plt.legend(loc='lower right')
```

```python
plt.plot(history.history['loss'], label='loss')
#plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='lower right')
plt.show()
```

# 5. Custom Model

```python
class MyModel(tf.keras.Model):
    def __init__(self):
        super(MyModel, self).__init__()
        self.conv1 = layers.Conv2D(6, 5, activation='relu')
        self.pool = layers.MaxPool2D(2, 2)
        self.conv2 = layers.Conv2D(16, 5, activation='relu')
        self.flatten = layers.Flatten()
        self.d1 = layers.Dense(120, activation='relu')
        self.d2 = layers.Dense(84, activation='relu')
        self.d3 = layers.Dense(K, activation='softmax')
    def call(self, x):
        x = self.pool(self.conv1(x))
        x = self.pool(self.conv2(x))
        x = self.flatten(x)
        x = self.d1(x)
        x = self.d2(x)
        x = self.d3(x)
        return x
```

```python
model2 = MyModel()
```

```python
model2.build(input_shape=(1,32,32,3))
```

```python
model2.summary(
    line_length=None,
    positions=None,
    print_fn=None,
    expand_nested=False,
    show_trainable=False,
)
```
